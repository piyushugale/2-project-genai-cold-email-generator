```python
# 2. Create a Chroma Client
import chromadb
chroma_client = chromadb.Client()

```

```python
# 3. Create a collection
# Collections are where you'll store your embeddings, documents, and any additional metadata.

# Delete the existing collection
# chroma_client.delete_collection(name="my_collection")
# Create a new collection
collection = chroma_client.create_collection(name="my_collection")
```

```python
# 4. Add some text documents to the collection

collection.add(
    ids=["id1", "id2"],
    documents=[
        "This is a document about scottish",
        "This is a document about apples"
    ]
)
```

```python
# Query the collection to find similar documents

import json

results = collection.query(
    query_texts=["This is a query document about glasgow"], # Chroma will embed this for you
    n_results=2 # how many results to return
)

print(results)
```
